```python
#plot a line plot
x = data['Sales']
y = data['Profit']
plt.plot(x, y, 'b')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('income')
plt.show()
```

```python
#histogram: plot a hostogram of sales
plt.hist(x)
plt.show()
```

```python
#quantitative count of observation of sales
sns.countplot(x='Sales', data= data)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='Sales', data=data)
```

```python
sns.lmplot(x='Sales', y='Profit', data=data)
```

```python
#plot a pairgrid of the store
sns.set_context('notebook', font_scale=1)
sns.set_style('darkgrid')
sns.PairGrid(data)
```

```python
#plot a scatter map with pairgrid
f = sns.PairGrid(data)
f.map(plt.scatter)
```

```python
f = sns.PairGrid(data)
f.map_diag(plt.hist)
f.map_upper(plt.scatter)
f.map_lower(plt.stackplot)
```
